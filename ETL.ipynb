```python
import pandas as pd

ruta_hechos = "C:\\Users\\licle\\Desktop\\PROYECTO INDIVIDUAL N°2\\homicidios.xlsx - HECHOS.csv"
ruta_victimas = "C:\\Users\\licle\\Desktop\\PROYECTO INDIVIDUAL N°2\\homicidios.xlsx - VICTIMAS.csv"

hechos = pd.read_csv(ruta_hechos)
victimas = pd.read_csv(ruta_victimas)
```

```python
#Revisamos datos númericos
#victimas.describe()

#Convertimos el tipo de dato de la columna edad 
victimas["EDAD"] = pd.to_numeric(victimas["EDAD"], errors='coerce')

victimas.describe()
```

```python
#eliminamos duplicados
victimas.drop_duplicates(inplace=True)
hechos.drop_duplicates(inplace=True)
print(victimas.shape)
print(hechos.shape)
#No existen
```

* En la descripción de los datos númericos notamos que los valores están dentro de los parametros normales por lo tanto concluimos en que no tenemos outliers


```python
#Eliminamos columnas redundantes o que no nos parecen relevantes

hechos_rev = hechos.drop(["Cruce","Altura","MM","DD", "HORA","XY (CABA)","Dirección Normalizada","PARTICIPANTES"], axis=1)
victimas_rev = victimas.drop(["AAAA","MM","DD", "VICTIMA"],axis=1)
```

```python
hechos_rev.head()
```

```python
victimas_rev.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Graficar subniveles de cada variable categórica
cols_cat1 = ["TIPO_DE_CALLE","COMUNA","Calle", "VICTIMA","ACUSADO"]


fig, ax = plt.subplots(nrows=len(cols_cat1),ncols =1, figsize=(10,30))
fig.subplots_adjust(hspace=1)

for i, col in enumerate(cols_cat1):
    sns.countplot(x=col, data=hechos_rev,ax=ax[i])
    ax[i].set_title(col)
    ax[i].set_xticklabels(ax[i].get_xticklabels(),rotation=30)
```

```python
cols_cat2 = ["ROL","SEXO"]
fig, ax = plt.subplots(nrows=len(cols_cat2),ncols =1, figsize=(10,30))
fig.subplots_adjust(hspace=1)

for i, col in enumerate(cols_cat2):
    sns.countplot(x=col, data= victimas_rev,ax=ax[i])
    ax[i].set_title(col)
    ax[i].set_xticklabels(ax[i].get_xticklabels(),rotation=30)
```

```python
#remplazamos PEATON_MOTO por PEATON para normalizar en un sólo subtipo de variable categórica
hechos_rev['VICTIMA'] = hechos_rev['VICTIMA'].replace('PEATON_MOTO', 'PEATON')
victimas_rev['ROL'] = victimas_rev['ROL'].replace('PASAJERO_ACOMPAÑANTE', 'PASAJERO')
```

```python
hechos_rev.to_csv("Hechos.csv", index=False)
victimas_rev.to_csv("Victimas.csv", index=False)
```
